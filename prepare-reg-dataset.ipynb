```python
from sklearn.model_selection import StratifiedKFold, KFold
import pickle

import os
import pandas as pds

dataname = 'house-prices'
df = pds.read_csv(os.path.join('data',dataname,'train.csv'))
```

```python
df = df.dropna(1).iloc[:,1:]

obj2int_dict = dict()
for key,col in df.dtypes.iteritems():
    if col == object: 
        obj2int_dict[key] = {v:n for n,v in enumerate(df[key].unique())}
        df[key] = df[key].apply(lambda x: obj2int_dict[key][x])
```

```python
X = df.drop(columns = 'SalePrice')
y = df.SalePrice

X.describe()
```

```python

ss1 = KFold(6, shuffle = True, random_state = 523)
ss2 = KFold(5, shuffle = True, random_state = 523)

for train_idx, test_idx in ss1.split(X,y):
    train_X, train_y = X.iloc[train_idx], y.iloc[train_idx]
    test_X, test_y = X.iloc[test_idx], y.iloc[test_idx]
    
    
    for train_idx, valid_idx in ss2.split(train_X,train_y):
        valid_X, valid_y = train_X.iloc[valid_idx], train_y.iloc[valid_idx]
        train_X, train_y = train_X.iloc[train_idx], train_y.iloc[train_idx]
        break
    break

with open(os.path.join('data',f'{dataname}.pkl'),'wb') as f:
    pickle.dump(
        dict(
            train_x = train_X.values,
            train_y = train_y.values,
            valid_x = valid_X.values,
            valid_y = valid_y.values,
            test_X = test_X.values,
            test_y = test_y.values,
            columns = train_X.columns,
        ),f
    )
```
